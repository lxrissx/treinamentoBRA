### 1. Ler e exibir um arquivo CSV usando pandas

```
import pandas as pd

df = spark.read.csv("dbfs:/FileStore/tables/retail.csv", header=True, sep=',')
df = df.toPandas()

df
```

### 2. Selecionar colunas específicas de um DataFrame usando pandas

```
df.filter(['stock_code', 'description', 'quantity', 'region'])
```

### 3. Filtrar linhas com base em uma condição usando pandas

```
df[(df['region'] == 'United Kingdom') & (df['quantity'] == '10')]
```

### 4. Agrupar dados com base em uma coluna usando pandas

```
df.agg({'invoice_date': ['min', 'max']})
```

```
top_10_invoice = (
    df
    .groupby('invoice_no')
    .agg(n        = ('stock_code', 'count'),
         datatime = ('invoice_date', 'first'))
    .reset_index()
    .sort_values('n', ascending=False)
    .head(10)
    .assign(date = date(2020, 3, 10))
)

top_10_invoice
```
