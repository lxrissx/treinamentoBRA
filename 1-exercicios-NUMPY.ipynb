### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
import numpy as np

arr = []
for a in range(10):
  arr.append(np.random.randint(50))

print(f"Valores aleatorios: {arr}")

media = np.mean(arr)
desvioP = np.std(arr)

print(f"Media: {media}")
print(f"Desvio padrão: {desvioP}")

#Solução do prof
arr.append(np.random.randint(50), size=10)
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
arr = np.random.rand(10)
  
print(f"Valores aleatorios: \n {arr} \n ")

for i in arr:
  if i <= 0.5:
    print(np.sqrt(i))

#Solução do prof
arr = np.random.rand(10)
arr[arr <= 0.5] = np.sqrt(arr[arr <= 0.5])
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
matriz = np.random.rand(3, 3, 3)
print(f"Matriz: \n {matriz} \n")

print("Maiores valores de cada linha")
print(np.max(matriz, axis=1))
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
matriz = np.random.rand(5, 5)
print(f"Valores aleatorios:\n {matriz}\n")

media = np.mean(matriz, axis=1) #axis=1 indica que queremos a media de cada linha da matriz
print(f"Media: {media}\n")

matrizMenosMedia = matriz - media[:, np.newaxis] #Primeiro transpomos o array que guarda as medias, depois subtraimos da matriz
print(f"Matriz Menos Media: \n {matrizMenosMedia}\n")
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
matriz = np.random.rand(4, 4) 

principal = np.diag(matriz) #Diag pega a diagonal da Matriz
secundaria = np.diag(np.fliplr(matriz)) #fliplr espelha a posição dos elementos mantendo os valores das linhas 
mediaDiagonais = np.mean([principal, secundaria]) # calcular a media com os valores das duas diagonais
desvioP = np.std([principal, secundaria]) # calcular o desvio padrão com os valores das duas diagonais
                     
print(f"Matriz:\n {matriz} \n")
print(f"Diagonal principal: {principal}")
print(f"Diagonal secundaria: {secundaria}")
print(f"Media Diagonais: {mediaDiagonais}")
print(f"desvio Padrao: {desvioP}")
```
