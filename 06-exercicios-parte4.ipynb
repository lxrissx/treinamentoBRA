### 21. Crie um programa em Python que receba uma string como entrada e imprima a string sem as vogais.

```python
texto = input("Digite uma frase: ")
vogais = ['a', 'e', 'i', 'o', 'u']
texto = list(texto)

for a in texto:
  if a in vogais:
    texto.pop(texto.index(a))

print(texto)
```

### 22. Crie um programa em Python que imprima a soma dos n primeiros números primos, onde n é um número fornecido pelo usuário.

```python
n = int(input("Digite um número: "))

i = 2
rodar = True

primos = []
while rodar:
    a = 2
    while a < i:
      primo = True
      if i % a == 0:
        primo = False
        break
      a += 1
    if(primo): primos.append(i)

    if len(primos) == n:
      rodar = False
    else:
      i += 1

print(primos)
#print(somaPrimos)
```

### 23. Crie um programa em Python que receba uma lista de números e imprima a soma dos números pares e a soma dos números ímpares presentes na lista.

```python
continuar = True
Lista = []

print("--------Digite os valores inteiros que deseja somar, para sair digite 'Exit'--------")
while continuar:
  elemento = input("Adicione um elemento a lista de números")
  
  try:
    elemento = int(elemento)
    Lista.append(elemento)
  except:
    continuar = False

somaPar = 0
somaImpar = 0

for a in Lista:
  if a % 2 == 0:
    somaPar += a
  else:
    somaImpar += a
    
print(f"Soma de números pares: {somaPar}")
print(f"Soma de números impares: {somaImpar}")
    
```

### 24. Crie um programa em Python que receba uma lista de strings e retorne uma lista com todas as strings que começam e terminam com a mesma letra.

```python
palavras = ["amor", "futebol", "verdade", "banana", "arara", "casa", "sol", "osso"]
palavras_selecionadas = [novaLista for novaLista in palavras if novaLista[0] == novaLista[-1]]

print(palavras_selecionadas)
```

### 25. Crie um programa em Python que leia um número do usuário e verifique se ele é um número perfeito. Um número perfeito é aquele que é igual à soma de seus divisores próprios (ou seja, a soma de seus fatores inteiros positivos, excluindo ele próprio).

```python
numero = int(input("Digite um número: "))

soma = 0
for a in range(1, numero-1):
  if numero % a == 0:
    soma += a

if soma == numero:
  print("É um número perfeito")
else:
  print("Não é um número perfeito")
```

### 26. Crie um programa em Python que calcule a série de Fibonacci para um determinado número de termos fornecido pelo usuário. A série de Fibonacci é uma sequência em que cada número é a soma dos dois números anteriores: 0, 1, 1, 2, 3, 5, 8, 13, ...

```python
termos = int(input("Digite o número de termos desejados: "))
fibonacci = [0, 1]
i = 2

while i < termos:
  valor = fibonacci[-1] + fibonacci[-2]
  fibonacci.append(valor)
  i += 1
  
print(fibonacci)
```
